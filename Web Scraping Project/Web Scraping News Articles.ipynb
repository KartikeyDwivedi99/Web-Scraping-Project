# Scraping Headlines and URL from "Times Of India"

```python
base_url1 = 'https://timesofindia.indiatimes.com/'
```

```python
r = requests.get(base_url1)
```

```python
r
```

```python
html_1 = r.content
```

```python
soup = BeautifulSoup(html_1,'lxml')
```

```python
soup
```

```python
headlines_1 = soup.find_all("figcaption")
headlines_1
```

```python
#extracting headlines

headline_1 = []
for headers in headlines_1:
    print(headers.text.strip())
    headline_1.append(headers.text.strip())
```

```python
#extracting URL's

url_1 = []
for link in soup.find_all('a', href=True):
    url_1.append(link['href'])
```

```python
url_1
```

```python
#removing irrelevant strings from href links

relevant_url = url_1[138:512]
relevant_url
```

```python
# removing duplicates from final_url using for loop

unique_url = []
for i in relevant_url:
    if i not in unique_url:
        unique_url.append(i)
  

print(unique_url)
```

# Scraping "Indian Express"

```python
#defining function to get headlines

def ind_exp_headlines(url):
    base_url_2 = url
    r = requests.get(base_url_2)
    html = r.content
    soup = BeautifulSoup(html, "lxml")
    title = soup.find_all("h2")
    headlines = []
    for titles in title:
        headlines.append(titles.text.strip())
    return headlines
```

```python
page_1_h = ind_exp_headlines('https://indianexpress.com/section/india/')
page_1_h
```
